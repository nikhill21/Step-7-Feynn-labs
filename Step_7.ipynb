```python
import numpy as np
import pandas as pd
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mcdonalds.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.plot(kind="bar",figsize=(90,20))
```

```python
crosstable = pd.crosstab(df['Gender'],df['VisitFrequency'])
crosstable
```

```python
from numpy.core.fromnumeric import size
props = {}
props[('Male','Every three months')] = {'facecolor':'red','edgecolor':'white'}
props[('Male','More than once a week')] = {'facecolor':'red','edgecolor':'white'}
props[('Male','Never')] = {'facecolor':'red','edgecolor':'white'}
props[('Male','Once a month')] = {'facecolor':'red','edgecolor':'white'}
props[('Male','Once a week')] = {'facecolor':'red','edgecolor':'white'}
props[('Male','Once a year')] = {'facecolor':'red','edgecolor':'white'}

props[('Female','Every three months')] = {'facecolor':'blue','edgecolor':'white'}
props[('Female','More than once a week')] = {'facecolor':'blue','edgecolor':'white'}
props[('Female','Never')] = {'facecolor':'blue','edgecolor':'white'}
props[('Female','Once a month')] = {'facecolor':'blue','edgecolor':'white'}
props[('Female','Once a week')] = {'facecolor':'blue','edgecolor':'white'}
props[('Female','Once a year')] = {'facecolor':'blue','edgecolor':'white'}

#plt.figure(figsize=(200,100))
mosaic(df,['Gender','VisitFrequency'],properties=props)
```
